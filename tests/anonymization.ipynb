```python
import glob
import re
import pandas as pd
import os

anon_codes = pd.read_csv("/data_isilon_main/isilon_images/10_MetaSystems/MetaSystemsData/Multimodal_Imaging_Daria/Publication/pseudoanonymization-codes.csv")
df = anon_codes[["Sample", "Sample_ID", "Tissue"]]

df['Tissue_Key'] = df['Tissue'].replace({'PT': 'TU', 'BM': 'BM'})
sample_id_map = pd.Series(df.Sample_ID.values, index=df.Sample + '_' + df.Tissue_Key).to_dict()

# Define the root directory containing the subfolders with files
root_dir = '/data_isilon_main/isilon_images/10_MetaSystems/MetaSystemsData/Multimodal_Imaging_Daria/Publication/processed_data'

def rename_entities(path):
    # Regex pattern to find "_<Sample>_BM" or "_<Sample>_TU"
    pattern = re.compile(r'(_)([^_]+)(_)(BM|TU)')
    
    for root, dirs, files in os.walk(path, topdown=False):
        # Rename files
        for name in files + dirs:
            old_path = os.path.join(root, name)
            match = pattern.search(name)
            if match:
                old_sample = match.group(2)  # The Sample part from the filename or directory name
                tissue_key = match.group(4)  # BM or TU
                map_key = f"{old_sample}_{tissue_key}"
                if map_key in sample_id_map:
                    new_sample_id = sample_id_map[map_key]
                    # Replace the old sample part with new sample ID
                    new_name = re.sub(pattern, r'\g<1>'+new_sample_id+r'\g<3>\g<4>', name)
                    new_path = os.path.join(root, new_name)
                    os.rename(old_path, new_path)
                    print(f'Renamed "{old_path}" to "{new_path}"')

rename_entities(root_dir)
```

```python

anon_codes = pd.read_csv("/data_isilon_main/isilon_images/10_MetaSystems/MetaSystemsData/Multimodal_Imaging_Daria/Publication/pseudoanonymization-codes.csv")

df = anon_codes[["Sample", "Sample_ID", "Tissue"]]

# Creating reverse mapping from Sample_ID to Sample
sample_to_id_map = pd.Series(df.Sample.values, index=df.Sample_ID).to_dict()

# Define the root directory containing the subfolders with files
root_dir = '/data_isilon_main/isilon_images/10_MetaSystems/MetaSystemsData/Multimodal_Imaging_Daria/Publication/test'

def rename_entities_back(path):
    # Regex pattern to identify the Sample_ID in the names
    pattern = re.compile(r'(_)([^_]+)(_)(BM|TU)')

    for root, dirs, files in os.walk(path, topdown=False):
        # Consider both files and directories for renaming
        for name in files + dirs:
            old_path = os.path.join(root, name)
            match = pattern.search(name)
            if match:
                sample_id = match.group(2)  # The Sample_ID from the filename or directory name
                if sample_id in sample_to_id_map:
                    original_sample = sample_to_id_map[sample_id]
                    # Replace the Sample_ID with the original sample part
                    new_name = re.sub(pattern, r'\g<1>'+original_sample+r'\g<3>\g<4>', name)
                    new_path = os.path.join(root, new_name)
                    os.rename(old_path, new_path)
                    print(f'Renamed "{old_path}" to "{new_path}"')

rename_entities_back(root_dir)
```
